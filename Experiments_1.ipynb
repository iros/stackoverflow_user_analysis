```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import time
import datetime as dt
import random
import scipy as sc
import scipy.stats as ss
import statsmodels.api as smapi
from statsmodels.formula.api import ols
import statsmodels.graphics as smgraphics 
from scipy.optimize import curve_fit

pd.set_option('notebook_repr_html', True)
pd.options.display.mpl_style = 'default'
rcParams['figure.figsize'] = 10, 10
```

```python
# ====== Create dataframe =======
tree = ET.parse('Users_sample.xml')
users = tree.getroot().getchildren()

# Create dataframe
def createTuples( users ):
    tuples = []
    counter = 0
    for user in users:         
        if user.get('Age'):      
            
            counter = counter + 1
                
            creationDate = dt.datetime.fromtimestamp(
                time.mktime(
                    time.strptime(user.get('CreationDate')[:-4], "%Y-%m-%dT%H:%M:%S")
                ))
            lastAccessDate = dt.datetime.fromtimestamp(
                time.mktime(
                    time.strptime(user.get('LastAccessDate')[:-4], "%Y-%m-%dT%H:%M:%S")
                ))
        
            rep = int(user.get('Reputation'))
            timeOnSite = (lastAccessDate - creationDate).days
            
            if (timeOnSite > 0):
                tuples.append(
                    (int(user.get('Age')), 
                     int(user.get('UpVotes')),
                     int(user.get('DownVotes')),
                     rep,
                     # '2008-08-26T00:16:53.810'
                     creationDate,
                     lastAccessDate,
                     timeOnSite,
                     (dt.datetime.now()-lastAccessDate).days,
                     rep / float(timeOnSite) )
                )
    print("Users found", counter, "out of", len(users))
    return tuples

def createDataFrame( users ):
    df = pd.DataFrame( createTuples( users ), 
                      columns=["Age", "UpVotes", "DownVotes", "Reputation", 
                               "CreationDate", "LastAccessDate", "TimeOnSite", "SinceLastAccess",
                               "RepOverTimeCoeff"])
    return df

df = createDataFrame( users )
```

```python
# Participating users are ones that have at least an upvote 
# or a downvote and whose reputation is not 1 (Which is the default.)
# Also remove 2 std of age
pu = df[df.UpVotes!=0][df.DownVotes!=0][df.Reputation!=1][df.Age <= 65]
print("percentage of remaining users", float(len(pu))/len(df))
```

```python
def gauss(x, *p):
    A, mu, sigma = p
    return A*numpy.exp(-(x-mu)**2/(2.*sigma**2))

hist, bin_edges = numpy.histogram(pu['Reputation'], density=True)
bin_centres = (bin_edges[:-1] + bin_edges[1:])/2

# p0 is the initial guess for the fitting coefficients (A, mu and sigma above)
p0 = [1., 0., 1.]

coeff, var_matrix = curve_fit(gauss, pu['Age'], pu['Reputation'], p0=p0)
print var_matrix
hist_fit = gauss(bin_centres, *coeff)

plt.plot(bin_centres, hist, label='Test data')
plt.plot(bin_centres, hist_fit, label='Fitted data')

print 'Fitted mean = ', coeff[1]
print 'Fitted standard deviation = ', coeff[2]

plt.show()
```

```python
# Trying to find a covarience function. The sample is too small. I don't think it's valid :(
# From: https://pythonhosted.org/infpy/gps.html
from numpy.random import seed
from infpy.gp import GaussianProcess, gp_1D_X_range
from infpy.gp import gp_plot_prediction, gp_learn_hyperparameters
from pylab import plot, savefig, title, close, figure, xlabel, ylabel
from infpy.gp import SquaredExponentialKernel as SE
from infpy.gp import noise_kernel as noise

#
# Part of X-space we are interested in
#
ageRepSum = []
for age in set(pu.Age):
    subset = pu[pu.Age==age]
    ageRepSum.append((age, min(subset.Reputation)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.25)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.5)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.75)))
    ageRepSum.append((age, max(subset.Reputation)))

ageRepSumDf = pd.DataFrame(ageRepSum, columns=["Age", "ReputationMedian"])

support = gp_1D_X_range(ageRepSumDf['Age'].min(), ageRepSumDf['Age'].max(), 1)

#
# Data
#
# sample = pu.ix[np.random.choice(pu.index.values, 200)]
sample = ageRepSumDf
X = list(sample['Age'].apply(lambda x: [x]))
Y = list(sample['ReputationMedian'])

def plot_gp(gp, fig_title, filename):
  figure()
  plot([x[0] for x in X], Y, 'ks', markersize=2, markerfacecolor='blue')
  mean, sigma, LL = gp.predict(support)
  gp_plot_prediction(support, mean, sigma)
  xlabel('Age')
  ylabel('ReputationMedian')
  title(fig_title)
  # savefig('%s.png' % filename)
  # savefig('%s.eps' % filename)
  
#
# Create a kernel with reasonable parameters and plot the GP predictions
#
kernel = SE([1.]) + noise(1.)
gp = GaussianProcess(X, Y, kernel)
plot_gp(
  gp=gp,
  fig_title='Initial parameters: kernel = SE([1]) + noise(1)',
  filename='learning_first_guess'
)

#
# Learn the covariance function's parameters and replot
#
gp_learn_hyperparameters(gp)
plot_gp(
  gp=gp,
  fig_title='Learnt parameters: kernel = SE([%.2f]) + noise(%.2f)' % (
    kernel.k1.params[0],
    kernel.k2.params.o2[0]
  ),
  filename='learning_learnt'
)
```
