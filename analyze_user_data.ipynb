## Defining participating users

One of the challenges of our dataset, is that we have a lot more users than we should be processing. To begin with, our Users.xml file contains 3,080,579 users, but only 373,188 have an age field, out of which only only 351,649 has an `Age` value that is actually set to a numeric value.

Out of those users, we need to choose "Participating Users" since we are trying to say something about
the generally active community of people on stack overflow. 

### What makes up a participating user?
For the purposes of our study, we define participating users as:
* Having at least an upvote or a downvote 
* and whose reputation is not 1 (Which is the default.)
* Also age < 65
* also user must have at least a question or an answer available to count as an active user.

We do not have user comments information at this point, but we could get them.

```python
pu = df[(df.UpVotes!=0)    & 
        (df.DownVotes!=0)  & 
        (df.Reputation!=1) & 
        (df.Age <= 65)     & 
        (
            (df.Questions > 0) | 
            (df.Answers > 0)   | 
            (df.Accepted_Answers > 0)
        )]
print("remaining users", len(pu), " out of " ,len(df))
print("percentage of remaining users", float(len(pu))/len(df))
```

What is the relationship between question asking and answering and reputation?
In this plot we select a subset minus 5% on each end of answers, questions and reputation and plot 
their mean per age.

Our Data contains:
* Answers (that were not accepted)
* Accepted_answers 
* Questions Asked

```python
# Change this function to look at different aggregation per age
# for example: median (Which looks similiar) or sum (which looks interesting...)
def plotSubset(subset, label):
    
    # To see what happens when we keep the outliers, uncomment this:
    # answer_subset = pu
    ageRepSum = []
    func = mean

    full_f_answers = func(pu.Answers)
    full_f_questions = func(pu.Questions)
    full_f_accepted_answers = func(pu.Accepted_Answers)
    full_f_reputation = func(pu.Reputation)

    # Build our dataset
    for age in set(subset.Age):
        age_subset = subset[subset.Age == age]
        ageRepSum.append((age, "Answers", func(age_subset.Answers) / full_f_answers))
        ageRepSum.append((age, "Accepted_Answers", func(age_subset.Accepted_Answers) / full_f_accepted_answers))
        ageRepSum.append((age, "Questions", func(age_subset.Questions) / full_f_questions))
        ageRepSum.append((age, "Reputation", func(age_subset.Reputation) / full_f_reputation))
    ageRepSumDf = pd.DataFrame(ageRepSum, columns=["Age", "Series", "Value"])


    fig = plt.figure()
    ax = fig.add_subplot(111) # single plot

    # Create subselections per available series
    a_subset = ageRepSumDf[ageRepSumDf.Series=="Answers"]
    aa_subset = ageRepSumDf[ageRepSumDf.Series=="Accepted_Answers"]
    q_subset = ageRepSumDf[ageRepSumDf.Series=="Questions"]
    r_subset = ageRepSumDf[ageRepSumDf.Series=="Reputation"]

    # Plot all series. By default not showing just answers, because there are so many
    # and they flatten the other graphs (but you can uncomment it.)

    # ax.plot(a_subset.Age, a_subset.Value, label="Answers", lw=2, color="#74D677")
    ax.plot(a_subset.Age, aa_subset.Value, label="Accepted_Answers", lw=2, color="#368A39")
    ax.plot(q_subset.Age, q_subset.Value, label="Questions", lw=2, color="#487BB5")
    ax.legend(loc='upper left')

    # Plot reputation on separate y axis on the right
    ax2 = ax.twinx()
    ax2.plot(r_subset.Age, r_subset.Value, label="Reputation", lw=2, ls=":", color='#CF740C')
    ax2.legend(loc='upper right')
    
    fig.suptitle(label, fontsize=20)

    plt.show()

# Get subset of data, removing %5 of each end.
answer_subset = pu[
    (pu.Answers > pu.Answers.quantile(q=0.05)) & 
    (pu.Answers < pu.Answers.quantile(q=0.95)) &
    (pu.Questions > pu.Questions.quantile(q=0.05)) & 
    (pu.Questions < pu.Questions.quantile(q=0.95)) &
    (pu.Reputation > pu.Reputation.quantile(q=0.05)) & 
    (pu.Reputation < pu.Reputation.quantile(q=0.95))
]

plotSubset(answer_subset, "90% of Answers, Questions and Reputation")
plotSubset(pu, "All data")
```

```python
def gauss(x, *p):
    A, mu, sigma = p
    return A*numpy.exp(-(x-mu)**2/(2.*sigma**2))

hist, bin_edges = numpy.histogram(pu['Reputation'], density=True)
bin_centres = (bin_edges[:-1] + bin_edges[1:])/2

# p0 is the initial guess for the fitting coefficients (A, mu and sigma above)
p0 = [1., 0., 1.]

coeff, var_matrix = curve_fit(gauss, pu['Age'], pu['Reputation'], p0=p0)
print var_matrix
hist_fit = gauss(bin_centres, *coeff)

plt.plot(bin_centres, hist, label='Test data')
plt.plot(bin_centres, hist_fit, label='Fitted data')

print 'Fitted mean = ', coeff[1]
print 'Fitted standard deviation = ', coeff[2]

plt.show()
```

```python
# Trying to find a covarience function. The sample is too small. I don't think it's valid :(
# From: https://pythonhosted.org/infpy/gps.html
from numpy.random import seed
from infpy.gp import GaussianProcess, gp_1D_X_range
from infpy.gp import gp_plot_prediction, gp_learn_hyperparameters
from pylab import plot, savefig, title, close, figure, xlabel, ylabel
from infpy.gp import SquaredExponentialKernel as SE
from infpy.gp import noise_kernel as noise

#
# Part of X-space we are interested in
#
ageRepSum = []
for age in set(pu.Age):
    subset = pu[pu.Age==age]
    ageRepSum.append((age, min(subset.Reputation)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.25)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.5)))
    ageRepSum.append((age, subset.Reputation.quantile(q=0.75)))
    ageRepSum.append((age, max(subset.Reputation)))

ageRepSumDf = pd.DataFrame(ageRepSum, columns=["Age", "ReputationMedian"])

support = gp_1D_X_range(ageRepSumDf['Age'].min(), ageRepSumDf['Age'].max(), 1)

#
# Data
#
# sample = pu.ix[np.random.choice(pu.index.values, 200)]
sample = ageRepSumDf
X = list(sample['Age'].apply(lambda x: [x]))
Y = list(sample['ReputationMedian'])

def plot_gp(gp, fig_title, filename):
  figure()
  plot([x[0] for x in X], Y, 'ks', markersize=2, markerfacecolor='blue')
  mean, sigma, LL = gp.predict(support)
  gp_plot_prediction(support, mean, sigma)
  xlabel('Age')
  ylabel('ReputationMedian')
  title(fig_title)
  # savefig('%s.png' % filename)
  # savefig('%s.eps' % filename)
  
#
# Create a kernel with reasonable parameters and plot the GP predictions
#
kernel = SE([1.]) + noise(1.)
gp = GaussianProcess(X, Y, kernel)
plot_gp(
  gp=gp,
  fig_title='Initial parameters: kernel = SE([1]) + noise(1)',
  filename='learning_first_guess'
)

#
# Learn the covariance function's parameters and replot
#
gp_learn_hyperparameters(gp)
plot_gp(
  gp=gp,
  fig_title='Learnt parameters: kernel = SE([%.2f]) + noise(%.2f)' % (
    kernel.k1.params[0],
    kernel.k2.params.o2[0]
  ),
  filename='learning_learnt'
)
```
