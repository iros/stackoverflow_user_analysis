# StackOverflow User Analysis

For those of you looking for an interesting dataset, StackOverflow offers a [data dump](http://blog.stackoverflow.com/2009/06/stack-overflow-creative-commons-data-dump/) of their entire user base, providing you with many GBs of play. We wanted to dig through this data, looking to understand what the community looks like.

In this article we will dive into the dataset, show you some patterns and learn some data analysis techniques along the way.

For those of you wanting to follow along, you can download our cleaned up data containing *user records with numeric age* [here](https://dl.dropboxusercontent.com/u/47185/all_user_data.csv).

````python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import time
import datetime as dt
import random
import scipy as sc
import scipy.stats as ss
import statsmodels.api as smapi
import statsmodels.graphics as smgraphics 
from statsmodels.formula.api import ols
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm

# Setup
pd.set_option('notebook_repr_html', True)
pd.options.display.mpl_style = 'default'
rcParams['figure.figsize'] = 10, 3

df = pd.DataFrame.from_csv('all_user_data.csv')
````

## What do we know about the users?

As of June 27th, 2014, Stackoverflow has 3,080,576 Users.

```
wc -l Users.xml
```

Out of those users, only 351,649 have a numeric Age setting. Which is approximatly 11.3% of the population.

````python
#removing xml tag + <users> root
total_users = 3080576 

age_breakdown = pd.DataFrame([[len(df), total_users - len(df)]], columns=["With age", "Without Age"])

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot(1,1,1)

age_breakdown.plot(kind="barh", stacked=True, sharey=True, ax=ax)
````

# Participating Users

Out of those 11.3%, we'd like to look only at the active users.
We define active users as such:

* They have a Reputation != 1 (which is the default)
* They have asked a question, provided an answer or had an answer accepted
* They have UpVotes or DownVotes
* They are below 70 years old 

This leaves us with 160,819 users out of our total, which is still 5.22% of the population.

We have also added a dimension to the data called `TimeOnSite`, which is the difference between a user's creation date and the last seen date. This way, if a user has become inactive a few years ago, we can take only that time frame into account.

````python
pu = df[(df.Reputation != 1) & 
        (df.Age <= 70) &
        ((df.UpVotes!=0) | (df.DownVotes!=0)) & 
        ((df.Questions > 0) | 
         (df.Answers > 0)   | 
         (df.Accepted_Answers > 0))]

print("remaining users", len(pu), " out of " ,total_users)
print("percentage of remaining users", float(len(pu))/total_users)

participating_breakdown = pd.DataFrame(
    [[len(pu), len(df) - len(pu), total_users - len(df)]], 
    columns=["Participating, With Age", "Not Participating, With Age", "Without Age"]
)

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot(1,1,1)

participating_breakdown.plot(kind="barh", stacked=True, sharey=True, ax=ax)
````

# What do we know about the participating users?

Out of our remaining users, the median age is 29 and the mean is 30, with a std of 7 years.

````python
ages = set(pu.Age)
print pu.Age.describe()
sns.distplot(pu.Age, kde=True, hist=True, bins=len(ages));
````

### They've been around for a while
They have spent an average of 989 days on the site, which is about 2 years and 8.5 months.

We see a pretty clear spike in users who have been on the site for > 2000 days.

````python
print pu.TimeOnSite.describe()
sns.distplot(pu.TimeOnSite, kde=True, hist=True, bins=100);
````

Looking at a joint distribution plot, we see that there is a slight correlation between programmers who have been on the site longer and them being older. This makes a lot of sense, in that, for you to be around on Stack Overflow longer, you probably need to have been older. 

````python
mpl.rc("figure", figsize=(6, 6))
sns.jointplot("Age", "TimeOnSite", pu, kind="kde");
````

Digging a bit further into the relationship between age and time on site, while we do see an upward trend (in that for ever year of age, you likely have spent about a month (27.7 days) longer on the site than someone one year younger than you,) we do not have a model that is accurate enough to predict that relationship. You can tell by just looking at the residuals and how large they are.

````python
# Sample random 5000 rows:
sample = pu.ix[np.random.choice(pu.index.values, 5000)]

x = 'Age'
y = 'TimeOnSite'

fig, ax = plt.subplots()

regression = ols("y ~ x", data=dict(x=sample[x], y=sample[y])).fit()
print regression.summary()

smgraphics.regressionplots.plot_fit(regression, 1, ax=ax)
smgraphics.regressionplots.abline_plot(model_results=regression, ax=ax)

````

# User Reputation

Looking at our user reputation, we can tell most of our users fall beneath our median of 189. Having said that, we have some prolific users who have managed to get as much as 671061 reputation points (if you're curious, see [Jon Skeet](http://stackoverflow.com/users/22656/jon-skeet))

We're going to reduce down our sample further, so that we can look at our users a bit more closely without the outliers skewing our sample too much. To do that, we'll take users 2 standard deviations from the mean. This still amounts to 99% of our users,

````python
mn = pu.Reputation.mean()
std = pu.Reputation.std()

subset = pu[(pu.Reputation > mn - 2 * std) & (pu.Reputation < mn + 2 * std)]
print len(subset) / float(len(pu))

print "Original Reputation \n{}".format(pu.Reputation.describe())
print "Subset Reputation \n{}".format(subset.Reputation.describe())
````

# Is there a relationship between age and reputation?

Looking at the linear regression between Age and Reputation, we see that while our model again is fairly bad, we have a slight upward trend. Specifically, for every year older, on average, a user has 30 more reputation points. Having said that, our model cannot be used to actually predict this.

````python
# Sample random 5000 rows:
sample = subset.ix[np.random.choice(subset.index.values, 5000)]

x = 'Age'
y = 'Reputation'

fig, ax = plt.subplots()

regression = ols("y ~ x", data=dict(x=sample[x], y=sample[y])).fit()
print regression.summary()

smgraphics.regressionplots.plot_fit(regression, 1, ax=ax)
smgraphics.regressionplots.abline_plot(model_results=regression, ax=ax)
````

````python
mpl.rc("figure", figsize=(6, 6))
sns.jointplot(subset["Age"], log(subset["Reputation"]), kind="kde")
````

````python
X = subset[['Age', 'TimeOnSite']]
y = subset['Reputation']

X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

print est.summary()

xx1, xx2 = np.meshgrid(np.linspace(X.Age.min(), X.Age.max(), 100), 
                       np.linspace(X.TimeOnSite.min(), X.TimeOnSite.max(), 100))

# plot the hyperplane by evaluating the parameters on the grid
Z = est.params[0] + est.params[1] * xx1 + est.params[2] * xx2

# create matplotlib 3d axes
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plot hyperplane
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plot data points - points over the HP are white, points below are black
# resid = y - est.predict(X)
# ax.scatter(X[resid >= 0].Age, X[resid >= 0].TimeOnSite, y[resid >= 0], color='black', alpha=0.4, facecolor='white')
# ax.scatter(X[resid < 0].Age, X[resid < 0].TimeOnSite, y[resid < 0], color='black', alpha=0.4)

# set axis labels
ax.set_xlabel('Age')
ax.set_ylabel('TimeOnSite')
ax.set_zlabel('Reputation')
````
